#  Visualization


#### 1. Stock prices fluctuate daily but follow long-term trends


```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], color='blue', linewidth=2)
plt.title("Stock Closing Price Trend", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Closing Price", fontsize=12)
plt.grid(True)
plt.show()
```

#### 2. Moving averages help identify trend direction


```python
df['MA7'] = df['Close'].rolling(window=7).mean()
df['MA30'] = df['Close'].rolling(window=30).mean()
plt.figure(figsize=(10,5))
plt.plot(df['Close'], label="Close Price")
plt.plot(df['MA7'], label="MA7 Day Moving Average")
plt.plot(df['MA30'], label="MA30 Day Moving Average")
plt.legend()
plt.title("Moving Average Trend")
plt.show()
```

---


#### 3. Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Stock Variables")
plt.show()
```

----

#### 4. Price Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Close'], bins=30, kde=True)
plt.title("Distribution of Closing Prices")
plt.show()
```

-----

#### 5. Box Plot (Outlier Detection)

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df['Close'])
plt.title("Closing Price Outlier Detection")
plt.show()
```

------

#### 6. RSI (Relative Strength Index)

```python
delta = df['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()
rs = avg_gain / avg_loss
df['RSI'] = 100 - (100 / (1 + rs))
plt.figure(figsize=(10,5))
plt.plot(df['RSI'])
plt.title("Relative Strength Index (RSI)")
plt.xlabel("Time")
plt.ylabel("RSI")
plt.show()
```

--------

#### 7. Bollinger Bands

```python
df['MA20'] = df['Close'].rolling(20).mean()
std = df['Close'].rolling(20).std()
df['Upper Band'] = df['MA20'] + (2 * std)
df['Lower Band'] = df['MA20'] - (2 * std)
plt.figure(figsize=(10,5))
plt.plot(df['Close'], label="Close Price")
plt.plot(df['MA20'], label="Moving Average")
plt.plot(df['Upper Band'], label="Upper Band")
plt.plot(df['Lower Band'], label="Lower Band")
plt.legend()
plt.title("Bollinger Bands")
plt.show()
```

------


#### 8. MACD (Moving Average Convergence Divergence)

MACD identifies trend changes and momentum shifts.

```python
exp1 = df['Close'].ewm(span=12, adjust=False).mean()
exp2 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = exp1 - exp2
df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
plt.figure(figsize=(10,5))
plt.plot(df['MACD'], label="MACD")
plt.plot(df['Signal'], label="Signal Line")
plt.legend()
plt.title("MACD Indicator")
plt.show()
```
